```python
import os
import pandas as pd
```

```python
path_prices = "Processed Datasets/prices_spy_qqq_2015_2025_stooq.csv"
path_sentiment = "Processed Datasets/2015-2025_sentiment.csv"
```

```python
prices = pd.read_csv(path_prices)
sentiment = pd.read_csv(path_sentiment)

prices["Date"] =  pd.to_datetime(prices["Date"], errors="coerce").dt.normalize()
sentiment["Date"] =  pd.to_datetime(sentiment["Date"], errors="coerce").dt.normalize()
```

```python
merged = prices.merge(sentiment, on="Date", how="left")
merged = merged[["Ticker","Date","Open","High","Low","Close","Volume","Sentiment","label_up_next"]]
merged
```

```python
os.makedirs("Processed Datasets", exist_ok=True)
out_merged = "Processed Datasets/merged_OHLCV_Sentiment.csv"
merged.to_csv(out_merged, index=False)
```
